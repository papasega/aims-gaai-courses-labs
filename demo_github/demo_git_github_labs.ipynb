# Demo: Using GitHub for Course Labs

This notebook shows the basic Git workflow students need for the GAAI-AIMS labs.

The goal is not to learn all of Git. The goal is to know how to:

- accept a GitHub Classroom assignment;
- clone the lab repository;
- edit notebooks and scripts;
- check modified files;
- commit work;
- push the final lab to GitHub;
- pull instructor updates when needed.

## 1. The GitHub Classroom Workflow

For each lab, the usual workflow is:

1. Accept the GitHub Classroom assignment link.
2. GitHub creates a personal repository for you.
3. Clone your personal repository on your computer or Colab environment.
4. Work on the notebook and scripts.
5. Commit your changes.
6. Push your work before the deadline.

Important: each student must push to their own repository, not to the instructor repository.

## 2. Configure Git Identity Once

Run this once on your machine, replacing the name and email with your own GitHub identity.

```bash
git config --global user.name "Your Name"
git config --global user.email "your_email@example.com"
```

Check the configuration:

```bash
git config --global --list
```

## 3. Clone Your Lab Repository

After accepting the GitHub Classroom link, copy the URL of your personal repository.

The URL usually looks like this:

```text
https://github.com/Generative-and-Agentic-AI-AIMS-classroom/lab-session-X-YOUR_USERNAME.git
```

Then clone it:

```bash
git clone https://github.com/Generative-and-Agentic-AI-AIMS-classroom/lab-session-X-YOUR_USERNAME.git
cd lab-session-X-YOUR_USERNAME
```

## 4. Enter the Lab Folder and Check Files

Example:

```bash
cd lab_s5
ls
```

Before editing anything, check the repository state:

```bash
git status
```

## 5. Work on the Lab

Typical files students may edit:

- `*.ipynb` notebooks;
- `*.py` utility files;
- `README.md` answers or observations;
- small result files requested by the instructor.

After editing, always run:

```bash
git status
```

Git will show modified files in red. These files are changed but not yet staged.

## 6. Stage Files with `git add`

To stage all modified files:

```bash
git add .
```

To stage one specific file:

```bash
git add lab_s5_prompting.ipynb
```

Then check again:

```bash
git status
```

Git will show staged files in green.

## 7. Commit Your Work

A commit is a saved checkpoint in Git.

Use a short but meaningful message:

```bash
git commit -m "lab_s5: complete prompting exercises"
```

Good commit messages:

- `lab_s5: complete zero-shot and few-shot prompts`
- `lab_s6: add mini-rag answers and evaluation`
- `lab_s7: complete qlora training observations`

Avoid vague messages such as:

- `update`
- `final`
- `changes`

## 8. Push to GitHub

After committing, push your work:

```bash
git push
```

If Git asks for authentication, use your GitHub account. Depending on your setup, GitHub may require:

- browser login;
- SSH key;
- personal access token for HTTPS.

After pushing, refresh the GitHub repository page and verify that your files appear online.

## 9. Full Submission Sequence

This is the sequence students will use most often:

```bash
git status
git add .
git commit -m "lab_s5: complete prompting exercises"
git push
git status
```

The final `git status` should say that the working tree is clean.

## 10. Pull Instructor Updates

If the instructor updates the lab repository or adds instructions, students may need to pull the latest changes.

Before pulling, save your work:

```bash
git status
git add .
git commit -m "save local work before pull"
```

Then pull:

```bash
git pull
```

If Git reports conflicts, do not panic. Open the conflicting files, keep the correct parts, then run:

```bash
git add .
git commit -m "resolve merge conflicts"
git push
```

## 11. Check the Remote Repository

Use this command to confirm where your work is pushed:

```bash
git remote -v
```

The URL should point to your GitHub Classroom repository.

## 12. Common Problems

### Problem: `nothing to commit`

Meaning: Git does not see any new changes. Either you already committed, or you saved the notebook in another folder.

### Problem: `failed to push some refs`

Meaning: the remote repository has changes that your local repository does not have. Usually run:

```bash
git pull
git push
```

### Problem: authentication failed

Meaning: GitHub did not accept your login method. Use browser login, SSH, or a GitHub personal access token depending on your setup.

### Problem: notebook output is very large

Clear unnecessary outputs before committing, especially large plots, model logs, or generated files.
